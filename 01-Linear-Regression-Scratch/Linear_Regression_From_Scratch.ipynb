```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('/content/Salary_dataset.csv')
data_cleaned = data.drop(columns = ['Unnamed: 0'])
print(data_cleaned)

def GradientDescent(m_now, b_now, points, L):
  m_gradient=0
  b_gradient=0

  n=len(points)

  for i in range(n):
    x=points.iloc[i].YearsExperience
    y=points.iloc[i].Salary

    m_gradient += -(2/n) * x * (y - m_now * x - b_now)
    b_gradient += -(2/n) * (y - m_now * x - b_now)

  m = m_now - L * m_gradient
  b = b_now - L * b_gradient

  return m,b

m = 0
b = 0
L = 0.0001
epochs = 10000

for i in range(epochs):
  m,b = GradientDescent(m, b, data_cleaned, L)
print(f'Bias = {m},Weight = {b}')

plt.scatter(data_cleaned.YearsExperience, data_cleaned.Salary, color = 'black', label='Data Points')

plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs. Years of Experience with Regression Line')

x_line = np.linspace(min(data_cleaned.YearsExperience), max(data_cleaned.YearsExperience), 100)
y_line = m * x_line + b
plt.plot(x_line, y_line, color='red', label='Regression Line')

plt.legend()
plt.show()
```
